```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df=pd.read_csv("/home/manuel/Documents/Bedu/proyecto-covid/datasets/paises.csv")
df.head()
```

```python
df_nona = df.dropna(subset = ['total_cases_per_million', 'hdi']) #para quitar na en columnas especificas
```

```python
df_nonaout = df_nona[df_nona.total_cases_per_million < 230000] #para quitar los outliers de esta variable
```

```python
df_nonaout.mean()
```

Creamos una nueva columna que divida al dataframe en dos grupos: 0 = se encuentra debajo o igual de la media de total de casos por millon, 1 = se encuentra por arriba de la media de total de casos por millon (53729.28)

```python
df_nonaout['arribaMediaTCPM'] = np.where(df_nonaout['total_cases_per_million'] > 53729.28, 1, 0)
df_nonaout.head()
```

```python
df_nonaout['abajoMediaTCPM'] = np.where(df_nonaout['total_cases_per_million'] <= 53729.28, 1, 0)
df_nonaout.head()
```

Realmente el Test AB es para aplicarse a un grupo control y un grupo donde se aplico la prueba. Este ejemplo es solo para fines didacticos, ya que nuestro grupo AB no tienen ningun control.

Ahora aniadimos el Test, que es si el pais esta arriba del HDI. Como el HDI toma de .7 a .79 como high human development, tomaremos como: A = hdi es igual o mayor a .7, B= hdi es menor a .7 (grupo control, aun que sabemos que esto no es un grupo control)

La Hipotesis es que, hay mayor cantidad de 0 en el grupo A, y mayor cantidad de 1 en el grupo B

```python
df_nonaout['hdiCategory'] = np.where(df_nonaout['hdi'] >= .7, 'A', 'B')
df_nonaout.tail()
```

Vamos a agruparlos para visualizar los conteos

```python
total_count = df_nonaout.groupby('hdiCategory')[['arribaMediaTCPM', 'abajoMediaTCPM']].sum()

total_count
```

```python
total_count['PorcentajeBajoMediaCasosTotales'] = total_count['abajoMediaTCPM'] / (total_count['abajoMediaTCPM'] + total_count['arribaMediaTCPM'])

total_count
```

Parece que nuestra hipotesis alternativa no es la correcta, y que la nula (hay mas paises con bajo HDI bajo la media de total de casos por millon) es la que es la correcta, y que pertenecear a un pais de HDI bajo es mejor, veamos que tan mejor

```python
diferencia_en_proporcion = 1 - (total_count.loc['A']['PorcentajeBajoMediaCasosTotales'] / total_count.loc['B']['PorcentajeBajoMediaCasosTotales'])
diferencia_en_porcentaje = diferencia_en_proporcion * 100
```

```python
print(f'El test B fue {diferencia_en_porcentaje:.2f}% más `efectivo que el test A para estar bajo el promedio de casos por millon.')
```

Vamos a aplicar un test de permutacion para ver si esto es estadisticamente significativo. Vamos a mezclar todos los resultados, pero sin la columna HDICategory para que no los podamos diferenciar

```python
all_measurements = df_nonaout[['arribaMediaTCPM', 'abajoMediaTCPM']]
all_measurements.head()
```

Ahora necesitamos saber cuántos tests con el flujo A hubieron para poder generar muestras del mismo tamaño exactamente:

```python
value_counts = df_nonaout['hdiCategory'].value_counts()

value_counts
```

Ahora vamos a aplicar el resto del algoritmo del test de permutación:

1. Usando muestreo aleatorio sin reposición, construye un nuevo grupo A del mismo tamaño que el original.
2. El resto de los datos conforman nuestro nuevo grupo B.
3. Cuantifica la métrica o estadística que calculaste con los grupos originales y guarda el resultado.
4. Repite los pasos 1-5 `R` veces para obtener una distribución de la estadística de interés.

```python
conv_a = []
no_conv_a = []
conv_b = []
no_conv_b = []

for _ in range(1000):
    a = df_nonaout.sample(value_counts.loc['A'], replace=False)
    conv_a.append(a['arribaMediaTCPM'].sum())
    no_conv_a.append(a['abajoMediaTCPM'].sum())
    
    b = df_nonaout.loc[~df_nonaout.index.isin(a.index)]
    conv_b.append(b['arribaMediaTCPM'].sum())
    no_conv_b.append(b['abajoMediaTCPM'].sum())
```

```python
perm_results = pd.DataFrame({
    'conv_a': conv_a,
    'no_conv_a': no_conv_a,
    'conv_b': conv_b,
    'no_conv_b': no_conv_b
})

perm_results
```

```python
perm_results['a_conv_rate'] = perm_results['conv_a'] / (perm_results['conv_a'] + perm_results['no_conv_a'])
perm_results['b_conv_rate'] = perm_results['conv_b'] / (perm_results['conv_b'] + perm_results['no_conv_b'])
```

```python
perm_results
```

```python
perm_results['diff'] = 1 - (perm_results['b_conv_rate'] / perm_results['a_conv_rate'])

perm_results
```
